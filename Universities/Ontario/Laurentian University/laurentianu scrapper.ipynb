```python
import pandas as pd
import ethnicolr as ec

df = pd.read_csv("data.csv", low_memory=False)
df[["First Name", "Last Name"]] = df["Name"].str.split(" ", n = 1, expand = True)
df.drop(columns =["Name"], inplace = True)
df = df[['First Name', 'Last Name', 'Email']]
df.fillna('No Email', inplace=True)
df = ec.pred_wiki_name(df,list(df)[1],list(df)[0])
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df)
df.to_csv("laurentian parsed.csv")
```
