```python
# import modules
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import Select
import pandas as pd
import ethnicolr as ec
from pathlib import Path
import re

# dictionary to hold data
data = {
    'First Name' : [],
    'Last Name' : [],
    'Email' : []
}

# open browser and go to directory
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--disable-extensions')
driver = str(Path().resolve().parents[2]) + "\Tools\chromedriver.exe"
browser = webdriver.Chrome(executable_path=driver, options=chrome_options)
timeout = 60
browser.get('https://ssbp.mycampus.ca/prod/www_directory.directory_uoit.p_main')

# find department selection box and search button
search_button = browser.find_element_by_xpath('/html/body/table/tbody/tr/td/table[1]/tbody/tr[8]/td[2]/input[1]')
departments_select = Select(browser.find_element_by_name('department_name_in'))
departments_labels = [o.text for o in departments_select.options]

# scrape all departments
for department_label in departments_labels:
    # select next department and click submit
    departments_select.select_by_visible_text(department_label)
    search_button.click()

    # find results
    results = browser.find_element_by_xpath('/html/body/table/tbody/tr/td/table').find_elements_by_tag_name('a')
    size = len(results)
    counter = 0
    
    # iterate through each person
    while (counter < size):
        prof = results[counter]
        prof.click()
        
        # obtain contact info
        full_name = browser.find_element_by_xpath('/html/body/table/tbody/tr/td/table/tbody/tr[4]/td[2]').text.split()
        fname = full_name[0]
        lname = full_name[-1]
        email = re.findall(r"[a-z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.[a-z]+", browser.page_source, re.I)
        if email != []:
            email = email[0]
        else:
            email = "No Email"
            
        # store contact info in data dictionary
        data['First Name'] += [fname]
        data['Last Name'] += [lname]
        data['Email'] += [email]

        # print contact info
        print(fname, lname, email)

        # go back to results page
        browser.execute_script("window.history.go(-1)")
        
        # find results
        results = browser.find_element_by_xpath('/html/body/table/tbody/tr/td/table').find_elements_by_tag_name('a')
        
        # increase counter by 1
        counter += 1
        
    # go back to directory page
    browser.execute_script("window.history.go(-1)")
    
    # find department selection box and search button
    search_button = browser.find_element_by_xpath('/html/body/table/tbody/tr/td/table[1]/tbody/tr[8]/td[2]/input[1]')
    departments_select = Select(browser.find_element_by_name('department_name_in'))
    departments_labels = [o.text for o in departments_select.options]
          
# create pandas dataframe and remove any duplicate entries
df = pd.DataFrame(data)
df = df.drop_duplicates()
    
# determine ethnicity
df = ec.pred_wiki_name(df,list(df)[1],list(df)[0])
    
# print dataframe and write data to csv file
print(df)
df.to_csv("uoit parsed.csv")
```
