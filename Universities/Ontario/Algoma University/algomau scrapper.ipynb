```python
# import modules
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import ethnicolr as ec
from pathlib import Path

# dictionary to hold data
data = {
    'First Name' : [],
    'Last Name' : [],
    'Email' : []
}

# initalize variables
full_name = None
fname = None
lname = None
email = None

# open browser and go to directory
response = requests.get('https://employees.algomau.ca/phonelist')

# use BeautifulSoup to parse raw code
soup=BeautifulSoup(response.text,'lxml')

# find results
results = soup.find("table", id="table").findAll("tr")

# iterate through each person
for prof in results:
    # obtain contact info
    if prof.find("td").get_text().split(", ") == ["Melissa Jones"]: # edge case
        fname = "Melissa"
        lname = "Jones"
        email = "Melissa.Jones@algomau.ca"
    elif prof.find("td").get_text().split(", ") == ["Susie Arbouw"]: # edge case
        fname = "Susie"
        lname = "Arbouw"
        email = "susie.arbouw@algomau.ca"
    else: # general case
        full_name = prof.find("td").get_text().split(", ")
        if len(full_name) == 1:
            continue
        fname = full_name[1].strip()
        lname = full_name[0].strip()
        email = re.findall(r"[a-z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.[a-z]+", str(prof), re.I)
        if email != []:
            email = email[0]
        else:
            email = "No Email"
    
    # store contact info in data dictionary
    data['First Name'] += [fname]
    data['Last Name'] += [lname]
    data['Email'] += [email]

    # print contact info
    print(fname, lname, email)

# create pandas dataframe and remove any duplicate entries
df = pd.DataFrame(data)
df = df.drop_duplicates()
    
# determine ethnicity
df = ec.pred_wiki_name(df,list(df)[1],list(df)[0])
    
# print dataframe and write data to csv file
print(df)
df.to_csv("algoma parsed.csv")
```
